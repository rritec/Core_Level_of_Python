    - float()

    - str()
    - bool()

## Converting int to float ?

```python
a_int = 11
```

```python
a_int
```

```python
type(a_int)
```

```python
a_int_to_float = float(a_int)
```

```python
a_int_to_float
```

```python
type(a_int_to_float)
```

## Converting int to complex ?

```python
a_int_to_complex = complex(a_int)
```

```python
a_int_to_complex
```

```python
type(a_int_to_complex)
```

## Converting int to str ?

```python
a_int_to_str = str(a_int)
```

```python
a_int_to_str
```

```python
type(a_int_to_str)
```

## Converting int to bool ?
> int 0 is equal to bool False


> int any other number is equal to bool True

```python
a_int_to_bool = bool(a_int)
```

```python
a_int_to_bool
```

```python
type(a_int_to_bool)
```
